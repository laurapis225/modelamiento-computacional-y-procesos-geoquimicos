# Ejercicios de Precipitación – Soluciones

## Ejercicio 1 – Crear listas a partir de datos

```python
# Lista con las abreviaturas de los meses (strings)
precip_months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
                 'jul', 'aug', 'sept', 'oct', 'nov', 'dec']

# Lista con los valores de precipitación en pulgadas (floats)
precip_inches = [0.70, 0.75, 1.85, 2.93, 3.05, 2.02,
                 1.93, 1.62, 1.84, 1.31, 1.39, 0.84]

print('Meses:', precip_months)
print('Precipitación (pulgadas):', precip_inches)
```

## Ejercicio 2 – Conversión de unidades en una lista

```python
# Factor de conversión: 1 pulgada = 25.4 mm
inch_to_mm = 25.4

# Copia de la lista original
precip_mm = precip_inches.copy()

# Convertir cada valor de pulgadas a milímetros
for i in range(len(precip_mm)):
    precip_mm[i] = precip_mm[i] * inch_to_mm

print('Precipitación (mm):', precip_mm)
print('Lista original sin modificar:', precip_inches)
```

## Ejercicio 3 – Crear una lista de listas

```python
# Agrupar ambas listas en una sola estructura
all_precip_data = [precip_months, precip_mm]

print(all_precip_data)
print('\nNúmero de sublistas:', len(all_precip_data))
print('Primera sublista (meses):', all_precip_data[0])
print('Segunda sublista (mm):', all_precip_data[1])
```

## Ejercicio 4 – Graficar datos desde una lista de listas

```python
import matplotlib.pyplot as plt

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar datos usando la lista de listas
ax.bar(
    all_precip_data[0],   # eje x: meses
    all_precip_data[1],   # eje y: precipitación en mm
    color='aqua'
)

# Configurar títulos y etiquetas
ax.set(
    title='Precipitación Media Mensual',
    xlabel='Mes',
    ylabel='Precipitación (mm)'
)

plt.show()
```
